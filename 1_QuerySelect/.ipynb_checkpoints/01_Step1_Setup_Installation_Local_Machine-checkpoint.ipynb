## 1.1.	Download the WaMDaM Wizard software    
Download the latest release from https://github.com/WamdamProject/WaMDaM_Wizard/releases
Choose a software option 

> **a.**	Save and run the WaMDaM Wizard executable “wamdam.exe” if you don’t have administrative access to the windows machine you are using. Click **Run** 

> **b.**	WaMDaM Wizard installer “WaMDaM_v1.02_UtahStateUniversity-1.02-amd64.msi” which can be installed and leaves a shortcut on your desktop. Click **More info** hyperlink you encounter a Windows warning dialog box, then click **Run anyway**

If you’re interested, the source code of the Wizard is available on GitHub here https://github.com/WamdamProject/WaMDaM_Wizard
<br>

## 1.2.	Launch WaMDaM Wizard     
Once installed, right click at the Wizard's shortcut and choose to run as an <font color=red>**Administrator**</font>. If you don’t run as Admin, some functionalities will not work.

Or double click at the executable and this main window will appear. Click **More info** hyperlink you encounter a Windows warning dialog box, then click **Run anyway**

## 1.3 Create a new, blank SQLite database file and connect to it
Click the **Connect to SQLite** tab (Figure 2), then click the button **Connect to a blank SQLite WaMDaM database** 

Provide the name of the database file (e.g., replicateWaMDaM) and specify where you want to store the file on your local machine (e.g., C:\desktop). The Wizard then will create a blank WaMDaM database with empty tables. You are ready to load data into the file.

Directions to deploy the WaMDaM database in three other relational database systems are available at https://github.com/WamdamProject/WaMDaM_Information_Model. _Note_, the wizard only works with SQLite database files.


    
## 1.4 Download the input datasets to load them into WaMDaM
Next, load data using one of three supported importers:   
i) from a WaMDaM Excel workbook template,  
ii) from the Consortium of Universities for the Advancement of Hydrologic Science, Inc. (CUAHSI) web-service, and  
iii) from the US Bureau Reclamation Water Information System (RWIS).    

Clone or download the use case data files from the **2Workbooks_datasets** folder in the following GitHub repository to a folder on your local machine.  

https://github.com/WamdamProject/WaMDaM_UseCases/tree/master/UseCases_files/2Workbooks_datasets
The data files are Excel macro template file (.xlsm) that is WaMDaM compliant and WaterML (.wml) from the Bureau of Reclamation Information System (Table 1). 

The Excel template must have 19 WaMDaM needed spreadsheets and their headers. You can download the blank template at following link. Open the template and view the HomePage spreadsheet for more info on its structure and how to populate it with data
https://github.com/WamdamProject/WaMDaM_UseCases/tree/master/UseCases_files/0WorkbookTemplates/InputData_Template

# 1.5 Set up a local Jupyter Notebook server on a local machine


###    Download or clone the GitHub repository of the Jupyter Notebook files to a folder on your machine.

https://github.com/WamdamProject/WaMDaM_JupyteNotebooks. For this paper, you only need to work with the files inside the folder called “1_QuerySelect”
Replace the existing already populated SQLite called BearRiverDatasets_June_2018_Final.sqlite with your SQLite file (e.g., replicateWaMDaM.sqlite). Next, you will set up the Jupyter Notebook Server with scripts that operate on your new SQLite file.   

### Install Jupyter Notebook Server for Python 2.7 (it requires some basic programing experience). 

Install Anaconda for Python 2.7 which already comes with the most common Python libraries are already installed. 
https://www.anaconda.com/download/  

For further details, follow instructions here http://jupyter.org/install  

The library called Potly which is used to plot the use case figures can be installed like this:


Run the Anaconda Command Prompt application, you can do so by clicking at the Windows start menu, then select Anaconda Prompt. Right click and <font color=red>**Run as Administrator**  </font> then type 

    conda install plotly 
    
If asked to Proceed ([y]/n)? type 

    y

## 1.6	View loaded data in WaMDaM tables (Optional)
This step here is not needed to replicate the work. If you want to view the WaMDaM table structure and populated data,

•	Download and install the free and open source tool **DB Browser For SQLite** to query the database and view its tables. Download from https://sqlitebrowser.org/
•	Download the already populated SQL **BearRiverDatasets_June_2018_Final.sqlite** file from GitHub at 
https://github.com/WamdamProject/WaMDaM_UseCases/tree/master/UseCases_files/3SQLite_database

•	Launch **DB Browser For SQLite** and Connect to the SQLite file you downloaded. Click Open Database. You can see the structure of WaMDaM tables by clicking at **Database Structure**. Click **Browse Data** to see the populated tables. Click **Execute SQL**. 
Type this simple query below and click at the execute triangle button. 

    SELECT * FROM ObjectTypes 

The query results in all the Object Type table columns with all its populated data in rows for all the Resource Types in the database.

# 1.7 Learn about controlled vocabularies (Optional)
This step is just for your information and in case you want to use it or make changes to the existing workbooks 

*  This step is also optional and not needed to replicate the work. Read further if you want to see how WaMDaM controlled vocabularies work or to suggesting new terms to add.   
<br>
*  WaMDaM controlled vocabulary are hosted online and can be accessed at http://vocabulary.wamdam.org/   
<br>

*  Each time you use the WaMDaM Wizard to load water management data, it calls this repository to download and update the SQLite controlled vocabulary tables.     
<br>
*  Continue to use your model’s native terms (e.g., how your model refers to object types, attributes, and instances). Add the controlled term next to each native term (i.e., register them against each other). Registering your model's native terms against these CVs will allow you to relate, query, and compare all your model’s data to other registered data from other models and datasets within the database. Open one of the Excel workbook examples to see how the CVs work.    

    In the HomePage spreadsheet, click at the button **Update Controlled Vocabularies** which will call the online WaMDaM controlled vocabulary registry and download or update the most recent vocabularies to your template.   

    Open the far right spreadsheet **ControlledVocabularies** which lists all of the downloaded terms for use as a dropdown menu in the rest of the input data spreadsheets.    
<br>
*  The source code for the vocabulary app can be accessed at https://github.com/WamdamProject/WaMDaM_ControlledVocabularies
